```python
#encode the categorial variables
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
categorial_col= ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose','Cuisines', 'Currency', 'Has Table booking', 'Has Online delivery','Is delivering now','Switch to order menu', 'Rating color', 'Rating text']
```

```python
for column in categorial_col:
  df[column]=label_encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
#Building Regression Model
x = df.drop(['Aggregate rating','Longitude', 'Latitude', 'Switch to order menu','Is delivering now','Locality Verbose'], axis=1)
y = df['Aggregate rating']
```

```python
print(x.shape)
print(y.shape)
```

```python
#training and testing the model
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr=LinearRegression()
```

```python
model = lr.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"MSE Linear Regression:{mse:.4f}")

print(f"R-squared Linear Regression: {r2:.4f}")
```

```python
#Linear Regression (OLS Model)
import statsmodels.api as sm
import numpy as np

x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
```

```python
#Decision Tree
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz

model_1 = DecisionTreeRegressor(min_samples_leaf=5)
result_1 = model_1.fit(x_train, y_train)
```

```python
y_pred_DR = result_1.predict(x_test)
mse_DR = mean_squared_error(y_test, y_pred_DR)
r2_Dr = r2_score(y_test, y_pred_DR)

print(f"MSE Decision Tree Regressor: {mse_DR:.4f}")
print(f"R-squared Decision Tree Regressor: {r2_Dr:.4f}")
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

model_2 = RandomForestRegressor()

result_2 = model_2.fit(x_train, y_train)

y_pred_Rf = result_2.predict(x_test)

mse_Rf = mean_squared_error(y_test, y_pred_Rf)
r2_Rf = r2_score(y_test, y_pred_Rf)

print(f"MSE Random Forest Regressor: {mse_Rf:.4f}")

print(f"R-squared Random Forest Regressor: {r2_Rf:.4f}")
```
